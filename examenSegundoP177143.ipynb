```python
import pandas as pd
```

```python
df = pd.read_csv("ventas.csv", parse_dates=["fecha"])

df = df.set_index("fecha")

print(df.head(3))

print(df.info())
```

```python
df = pd.read_csv("items.csv")
df = df.eval("importe = precio * cantidad")
dfa = df.query("categoria == 'A'")
print(df.groupby("categoria")["importe"].mean())
```

```python
dc = pd.read_csv("clientes.csv")
do = pd.read_csv("ordenes.csv")
```

```python
merge = dc.merge(do, on="cliente_id", how="left")
print(len(merge))
print(merge.groupby("cliente_id")["monto"].sum().fillna(0))
```

```python
df = pd.read_csv("sensores.csv")[["ts", "temperatura", "humedad"]]
df["ts"] = pd.to_datetime(df["ts"], format="mixed", errors="coerce")
df["temperatura"] = df["temperatura"].fillna(df["temperatura"].median())
df["humedad"] = df["humedad"].fillna(df["humedad"].mean())
```

```python
print(df["ts"].isna().sum())
print(df["temperatura"].median())
print(df["humedad"].mean())
```
